```python
# ìƒí’ˆ ì¬ê³  ê´€ë¦¬ ì‹œìŠ¤í…œ í´ë˜ìŠ¤
class InventoryManager:
    """ìƒí’ˆ ì¬ê³ ë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self):
        """ì´ˆê¸° ì¬ê³  ë”•ì…”ë„ˆë¦¬ ìƒì„±"""
        self.inventory = {
            "ì‚¬ê³¼": 10,
            "ë°”ë‚˜ë‚˜": 5,
            "ì²´ë¦¬": 8
        }
        print("ğŸ“¦ ì¬ê³  ê´€ë¦¬ ì‹œìŠ¤í…œì´ ì‹œì‘ë˜ì—ˆìŠµë‹ˆë‹¤.")
        print(f"ì´ˆê¸° ì¬ê³ : {self.inventory}")
    
    def add_product(self, product_name, quantity):
        """ìƒˆë¡œìš´ ìƒí’ˆ ì¶”ê°€"""
        if product_name in self.inventory:
            print(f"âš ï¸ '{product_name}'ì€(ëŠ”) ì´ë¯¸ ì¡´ì¬í•©ë‹ˆë‹¤. ì¬ê³ ë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.")
            self.inventory[product_name] += quantity
        else:
            self.inventory[product_name] = quantity
            print(f"âœ… '{product_name}' {quantity}ê°œê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")
    
    def update_quantity(self, product_name, new_quantity):
        """íŠ¹ì • ìƒí’ˆì˜ ì¬ê³  ë³€ê²½"""
        if product_name in self.inventory:
            old_quantity = self.inventory[product_name]
            self.inventory[product_name] = new_quantity
            print(f"âœ… '{product_name}' ì¬ê³ ê°€ {old_quantity}ê°œì—ì„œ {new_quantity}ê°œë¡œ ë³€ê²½ë˜ì—ˆìŠµë‹ˆë‹¤.")
        else:
            print(f"âŒ '{product_name}'ì€(ëŠ”) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
    
    def remove_product(self, product_name):
        """íŠ¹ì • ìƒí’ˆ ì‚­ì œ"""
        if product_name in self.inventory:
            quantity = self.inventory.pop(product_name)
            print(f"âœ… '{product_name}' {quantity}ê°œê°€ ì‚­ì œë˜ì—ˆìŠµë‹ˆë‹¤.")
        else:
            print(f"âŒ '{product_name}'ì€(ëŠ”) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
    
    def show_all_products(self):
        """ì „ì²´ ìƒí’ˆ ëª©ë¡ ì¡°íšŒ"""
        print("\nğŸ“‹ ì „ì²´ ìƒí’ˆ ëª©ë¡:")
        print("-" * 30)
        for product, quantity in self.inventory.items():
            print(f"  {product}: {quantity}ê°œ")
        print("-" * 30)
        print(f"ì´ {len(self.inventory)}ê°œ ìƒí’ˆ")
    
    def search_product(self, product_name):
        """íŠ¹ì • ìƒí’ˆ ê²€ìƒ‰"""
        if product_name in self.inventory:
            quantity = self.inventory[product_name]
            print(f"ğŸ” '{product_name}' ê²€ìƒ‰ ê²°ê³¼: {quantity}ê°œ")
        else:
            print(f"âŒ '{product_name}'ì€(ëŠ”) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
    
    def get_total_products(self):
        """ì´ ìƒí’ˆ ìˆ˜ ë°˜í™˜"""
        return len(self.inventory)
    
    def get_total_quantity(self):
        """ì´ ì¬ê³  ìˆ˜ëŸ‰ ë°˜í™˜"""
        return sum(self.inventory.values())
```

```python
# ë©”ì¸ ì‹¤í–‰ ë° í…ŒìŠ¤íŠ¸
print("=" * 60)
print(" Task 1: ìƒí’ˆ ì¬ê³  ê´€ë¦¬ (ë”•ì…”ë„ˆë¦¬ í™œìš©)")
print("=" * 60)

# 1. ì¬ê³  ê´€ë¦¬ ì‹œìŠ¤í…œ ìƒì„±
manager = InventoryManager()

print("\n" + "=" * 40)
print("1ï¸âƒ£ ìƒˆë¡œìš´ ìƒí’ˆ ì¶”ê°€ í…ŒìŠ¤íŠ¸")
print("=" * 40)

# ìƒˆë¡œìš´ ìƒí’ˆ ì¶”ê°€
manager.add_product("ì˜¤ë Œì§€", 15)
manager.add_product("í¬ë„", 7)
manager.add_product("ì‚¬ê³¼", 5)  # ê¸°ì¡´ ìƒí’ˆì— ì¶”ê°€

print("\n" + "=" * 40)
print("2ï¸âƒ£ íŠ¹ì • ìƒí’ˆì˜ ì¬ê³  ë³€ê²½ í…ŒìŠ¤íŠ¸")
print("=" * 40)

# ì¬ê³  ë³€ê²½
manager.update_quantity("ë°”ë‚˜ë‚˜", 12)
manager.update_quantity("ì—†ëŠ”ìƒí’ˆ", 10)  # ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ìƒí’ˆ

print("\n" + "=" * 40)
print("3ï¸âƒ£ ì „ì²´ ìƒí’ˆ ëª©ë¡ ì¡°íšŒ í…ŒìŠ¤íŠ¸")
print("=" * 40)

# ì „ì²´ ëª©ë¡ ì¡°íšŒ
manager.show_all_products()

print("\n" + "=" * 40)
print("4ï¸âƒ£ íŠ¹ì • ìƒí’ˆ ê²€ìƒ‰ í…ŒìŠ¤íŠ¸")
print("=" * 40)

# ìƒí’ˆ ê²€ìƒ‰
manager.search_product("ì²´ë¦¬")
manager.search_product("í¬ë„")
manager.search_product("ì—†ëŠ”ìƒí’ˆ")

print("\n" + "=" * 40)
print("5ï¸âƒ£ íŠ¹ì • ìƒí’ˆ ì‚­ì œ í…ŒìŠ¤íŠ¸")
print("=" * 40)

# ìƒí’ˆ ì‚­ì œ
manager.remove_product("ì²´ë¦¬")
manager.remove_product("ì—†ëŠ”ìƒí’ˆ")

print("\n" + "=" * 40)
print("6ï¸âƒ£ ìµœì¢… ì¬ê³  í˜„í™©")
print("=" * 40)

# ìµœì¢… í˜„í™©
manager.show_all_products()
print(f"\nğŸ“Š ì´ ìƒí’ˆ ìˆ˜: {manager.get_total_products()}ê°œ")
print(f"ğŸ“Š ì´ ì¬ê³  ìˆ˜ëŸ‰: {manager.get_total_quantity()}ê°œ")

print(" Task 1 ì™„ë£Œ!")
```
