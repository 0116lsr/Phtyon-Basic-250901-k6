```python
#deque , ê° ìš”ì†Œì˜ ë‹¤ìŒ ìš”ì†Œì˜ ë©”ëª¨ë¦¬ ìœ„ì¹˜ ì£¼ì†Œê°’ì„ ì•Œê³  ìˆë‹¤.
# ë§ˆì§€ë§‰ ìš”ì†ŒëŠ” , ë§¨ ì²˜ìŒ ìš”ì†Œì˜ ë©”ëª¨ë¦¬ ìœ„ì¹˜ ì£¼ì†Œê°’ì„ ì•Œê³  ìˆë‹¤.
# ê²°ë¡ , ì²˜ìŒê³¼ ëì´ ì—°ê²°ì´ ëœêµ¬ì¡°ê°€ ë˜ê³ , ì´ê²ƒì„ ì´ìš©í•´ì„œ,
# ê° ìš”ì†Œë¥¼ ìˆœíšŒ ì‹œí‚¬ìˆ˜ ìˆë‹¤.
# rotate() í•¨ìˆ˜ë¥¼ ì´ìš©í•´ì„œ, ê° ìš”ì†Œë¥¼ ì´ë™ì´ ê°€ëŠ¥í•˜ë‹¤.
#deque , íì²˜ëŸ¼ ì‚¬ìš©í•˜ê¸°, appendleft() í•¨ìˆ˜ ì´ìš©í•˜ê¸°.
# ì™¼ìª½ë¶€í„° ì…ë ¥í•˜ê²Œ í•´ì„œ, ë“¤ì–´ê°„ ê°’ë¶€í„° ì¶œë ¥í•˜ê¸°. FIFO
from collections import deque

deque_list = deque()

 # 0 ~ 4, range(ì‹œì‘,ë, ì¦ê°€ê°’), ëê°’ì€ ë¯¸í¬í•¨
for i in range(5):
  deque_list.appendleft(i)
# ê¸°ì¡´ì— append ë¡œ ì¶”ê°€ì‹œ, ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì¶”ê°€í–ˆìŒ.
# [ 0,1,2,3,4]
print(f"deque_list ì¶œë ¥ : {deque_list}")
# deque_list ì¶œë ¥ : deque([4, 3, 2, 1, 0])

# rotate() í•¨ìˆ˜ ì´ìš©
# 2ì¹¸ ì´ë™í•´ë³´ê¸°.
deque_list.rotate(2)
print(f"deque_list ì¶œë ¥ : {deque_list}")
```

íŒŒì´ì¬ì˜ dequeëŠ” 'double-ended queue'ì˜ ì•½ìë¡œ,

ì£¼ìš” ìš©ë„:

í(Queue): ë¨¼ì € ë“¤ì–´ì˜¨ ë°ì´í„°ê°€ ë¨¼ì € ë‚˜ê°€ëŠ”(FIFO) êµ¬ì¡°ë¥¼ êµ¬í˜„í•  ë•Œ.

```python
# reversed() í™•ì¸.
from collections import deque

deque_list = deque()

 # 0 ~ 4, range(ì‹œì‘,ë, ì¦ê°€ê°’), ëê°’ì€ ë¯¸í¬í•¨
for i in range(5):
  deque_list.appendleft(i)
# ê¸°ì¡´ì— append ë¡œ ì¶”ê°€ì‹œ, ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì¶”ê°€í–ˆìŒ.
# [ 0,1,2,3,4]
print(f"deque_list ì¶œë ¥ : {deque_list}")
# deque_list ì¶œë ¥ : deque([4, 3, 2, 1, 0])
deque_list_reversed = deque(reversed(deque_list))
print(f"deque_list ì¶œë ¥ : {deque_list_reversed}")
```

OrderedDictëŠ” ì´ë¦„ ê·¸ëŒ€ë¡œ **ìˆœì„œê°€ ë³´ì¥ë˜ëŠ” ë”•ì…”ë„ˆë¦¬(Dictionary)**ì…ë‹ˆë‹¤.

```python
# ì •ë ¬ í•´ë³´ê¸°.
# OrderdDict , ìˆœì„œìˆëŠ” ë”•ì…˜ë„ˆë¦¬ ì´ìš©í•˜ê¸°.
def sort_by_key(t):
  return t[0]

from collections import OrderedDict
d= OrderedDict()
d["x"] = 1
d["b"] = 2
d["d"] = 4
d["c"] = 3
d["e"] = 5

# key ì •ë ¬í•´ë³´ê¸°.
for k,v in OrderedDict(sorted(d.items(), key=sort_by_key)).items():
  print(f"key : {k}, value : {v}")
```

```python
from collections import Counter

sentence = "íŒŒì´ì¬ íŒŒì´ì¬ ì •ë§ ì •ë§ ì •ë§ ì¬ë¯¸ìˆë‹¤ íŒŒì´ì¬ ìµœê³ "
words = sentence.split() # ë¬¸ì¥ì„ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¡œ ë¶„ë¦¬
print(words)

word_counts = Counter(words)

# ê°€ì¥ í”í•œ ë‹¨ì–´ 3ê°œ ì°¾ê¸°
top_3_words = word_counts.most_common(3)
print(top_3_words)
```
